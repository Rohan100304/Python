UNIFORM DISTRIBUTION

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the Uniform Distribution
a = 0  # minimum value
b = 10  # maximum value

# Create a sample of random numbers following a Uniform Distribution
sample_size = 1000
samples = np.random.uniform(a, b, sample_size)
samples
```

```python
# Plot the histogram of the samples to visualize the distribution
plt.hist(samples, bins=30, density=True, alpha=0.4, color='g')

# Add the theoretical Uniform distribution's PDF
x = np.linspace(a, b, 1000)
pdf = np.ones_like(x) / (b - a)  # Uniform distribution PDF

plt.plot(x, pdf, 'r-', label="Uniform PDF")
plt.title('Uniform Distribution (a=0, b=10)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

NORMAL DISTRIBUTION

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define mean and standard deviation
mean = 0
std_dev = 4


```

```python
# Generate values for the x-axis (from -5 to 5)
x = np.linspace(-5, 5, 1000)

# Calculate the normal distribution PDF
pdf = norm.pdf(x, mean, std_dev)
pdf
```

```python
# Plot the PDF
plt.plot(x, pdf, label=f'Normal Distribution\nMean={mean}, Std={std_dev}')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the range for which we want to calculate the probability
a = -1  # Lower bound
b = 1   # Upper bound

# Calculate the probability of falling between a and b
probability = norm.cdf(b, mean, std_dev) - norm.cdf(a, mean, std_dev)

# Print the probability
print(f"The probability of a value falling between {a} and {b} is: {probability:.4f}")
```

EXPONENTIAL DISTRIBUTION

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Defining rate parameter (lambda)
lambda_param = 1 

# Generate values for the x-axis (with range 0 to 5)
x = np.linspace(0, 5, 1000)

x
```

```python
# Calculate the exponential distribution PDF
pdf = expon.pdf(x, scale=1/lambda_param)

# Plot the PDF
plt.plot(x, pdf, label=f'Exponential Distribution\nLambda={lambda_param}')
plt.title('Exponential Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

GAMMA DISTRIBUTION

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parameters for the Gamma distribution
shape = 2  # Shape parameter (k or α)
scale = 2  # Scale parameter (θ)

# Generate x values
x = np.linspace(0, 20, 1000)


```

```python
# Compute the Gamma probability density function (PDF)
pdf = gamma.pdf(x, a=shape, scale=scale)

pdf
```

```python
# Plot the distribution
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label=f'Gamma PDF (shape={shape}, scale={scale})', color='blue')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Gamma Probability Distribution')
plt.legend()
plt.grid()
plt.show()
```
