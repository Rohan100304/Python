```python
#Step 3: Data Preprocessing
X = data.iloc[:,:-1].values  #Select all rows and all columns except last one
y = data.iloc[:,1].values  #All rows and second column
```

```python
#Step 4: Splitting the Dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Step 5:Fitting linear regression model into the training set
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train) #actually produces the linear eqn for the data
```

```python
y_pred = model.predict(X_test) 
y_pred
```

```python
y_test
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual Data')
plt.plot(X_test, y_pred, color='red', linewidth=3, label='Linear Regression Prediction')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# Evaluation metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
#coefficients and evaluation metrics
print("Coefficient (slope):", model.coef_[0])
print("Intercept:", model.intercept_)
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```
