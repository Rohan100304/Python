```python
#STEP 1: LOAD THE IRIS DATA SET
import pandas as pd

data = pd.read_csv(r"C:\Users\Rohan\Downloads\Iris.csv")
data
```

```python
#STEP 2: STANDARDIZE THE DATA

from sklearn.preprocessing import StandardScaler
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

```python
# Separating features and target
x = data.loc[:, features].values
y = data.loc[:,['Species']].values
```

```python
# Standardizing the features(the result is stored in x)
x = StandardScaler().fit_transform(x)
```

```python
# Convert to DataFrame for better visualization
df1 = pd.DataFrame(x)
df1
```

```python
# Verify mean and standard deviation
print("Mean of each feature after standardization:")
print(df1.mean().round(6))  # Should be approximately 0

#print("\nStandard deviation of each feature after standardization:")
print(df1.std(ddof=0).round(6))  # ddof=0 ensures that the formula matches what StandardScaler
```

```python
STEP 3: FIT PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)    #Giving PCA with No. of components
pca
```

```python
#Performing PCA Transformation
principalComponents = pca.fit_transform(x)
principalComponents
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC 1', 'PC 2'])
principalDf        #Contains PC
```

```python
finalDf = pd.concat([principalDf, data[['Species']]], axis = 1)    
finalDf 
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Species'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'PC 1']
               , finalDf.loc[indicesToKeep, 'PC 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
